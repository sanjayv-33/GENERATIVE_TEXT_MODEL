```python
from transformers import pipeline
import textwrap

# 1. Setup the generator
generator = pipeline('text-generation', model='gpt2')

# 2. Get your prompt
my_prompt = input("Enter your topic: ")

# 3. Generate text
print("\n" + "="*50)
print(f"PROMPT: {my_prompt}")
print("="*50 + "\n")

output = generator(
    my_prompt,
    max_length=200,
    num_return_sequences=1,
    truncation=True,
    pad_token_id=50256,
    temperature=0.8,
    no_repeat_ngram_size=2
)

# 4. Format the output for readability
raw_text = output[0]['generated_text']

# wrap text at 80 characters wide
wrapped_text = textwrap.fill(raw_text, width=80)

print(wrapped_text)
print("\n" + "="*50)
```
